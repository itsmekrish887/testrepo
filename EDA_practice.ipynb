```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
data = {
    'feature1' :[1,2,np.nan, 4, 5],
    'feature2' :[6,np.nan, 8,9,10],
    'target' :[1,2,3,4,5]
}
df = pd.DataFrame(data)
df
```

```python
for column in df.columns:
    if df[column].isnull().any():

        # Separate the data into training and testing sets
        known_data = df[df[column].notnull()]
        unknown_data = df[df[column].isnull()]

        x_train = known_data.drop(column, axis=1)
        y_train = known_data[column]

        x_test = unknown_data.drop(column, axis=1)

        # Combine x_train and y_train to remove any null rows
        train_data = pd.concat([x_train, y_train], axis=1).dropna()
        x_train = train_data.drop(column, axis=1)
        y_train = train_data[column]

        # Train model
        LR = LinearRegression()
        LR.fit(x_train, y_train)

        # Predict missing values
        predict_values = LR.predict(x_test)

        # Replace null values with predicted values
        df.loc[df[column].isnull(), column] = predict_values
```

#How to remove outliers from all numerical columns using a function

```python
df = pd.read_csv('/content/data.csv')
```

```python
df.head()
```

Remove outliers based on the below conditions

*   if any value is above (Q3 + 1.5IQR)  or lower than (Q1 - 1.5*IQR)



```python
Q1 = df['Engine HP'].quantile(0.25)
Q3 = df['Engine HP'].quantile(0.75)
IQR = Q3-Q1
```

```python
# Calculate the upper and lower bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to remove outliers
df_no_outliers = df[~((df['Engine HP'] < lower_bound) | (df['Engine HP'] > upper_bound))]

```
